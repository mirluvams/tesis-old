# Arquitectura del modelo

La arquitectura **Swin Transformer** combina las caractersticas de los transformadores y las redes convolucionales para el procesamiento eficiente y efectivo de imgenes. 

**Componentes clave de la arquitectura Swin Transformer:**

* **Divisin de la imagen en parches**: La imagen de entrada se divide en parches de tamao fijo (por ejemplo, 4x4 pxeles). Esto permite al modelo capturar informacin contextual a nivel de parche y aprender relaciones espaciales entre ellos.

* **Transformador de ventanas**: A diferencia de los transformadores tradicionales que operan a nivel de tokens secuenciales, Swin Transformer utiliza un transformador de ventanas. Se agrupan los parches en ventanas ms grandes y se aplican las operaciones de atencin entre las ventanas. Esto ayuda a capturar relaciones a largo plazo entre los parches.

* **Capas de transformador**: La arquitectura Swin Transformer incluye mltiples capas de transformador, donde cada capa tiene bloques de transformador. Cada bloque de transformador contiene una capa de atencin multi-cabeza para capturar relaciones entre los parches y una capa de feed-forward para procesar la informacin dentro de cada parche.

* **Reduccin de escala**: Dado que Swin Transformer opera en ventanas a nivel de parches, la resolucin espacial se reduce a medida que se profundiza en la red. Para preservar la informacin espacial, se utiliza una operacin de reduccin de escala en las capas para fusionar las caractersticas de las ventanas adyacentes.

* **Clasificador final**: Despus de varias capas de transformador, se aplica un clasificador final para generar las predicciones de salida. Este clasificador puede variar segn la tarea especfica, como la clasificacin de imgenes en mltiples clases.




**Diferencia entre dos arquitecturas "populares" para clasificar imgenes con transformers:**

**Estructura de entrada**: En ViT, la imagen de entrada se divide en parches y se considera una secuencia 1D de parches. Cada parche se trata como un token y se procesa utilizando capas de atencin. En cambio, Swin Transformer divide la imagen en parches, pero los agrupa en ventanas ms grandes y aplica atencin entre las ventanas. Esto permite que Swin Transformer capture relaciones a largo plazo entre los parches.

**Atencin**: En ViT, se utiliza la atencin de escala global, donde cada token se relaciona con todos los dems tokens de la secuencia. Esto significa que la atencin se aplica a nivel de token individual. En Swin Transformer, se utiliza una atencin a nivel de ventana, donde se aplica atencin entre las ventanas agrupadas de parches. Esto reduce la complejidad computacional al limitar el alcance de la atencin.

**Escala espacial**: En ViT, la escala espacial de la imagen se pierde a medida que se dividen en parches y se consideran como una secuencia 1D. Para compensar esto, se agrega informacin posicional a los tokens para ayudar al modelo a comprender la estructura espacial de la imagen. En Swin Transformer, se utiliza una reduccin de escala para fusionar caractersticas en las ventanas adyacentes y preservar informacin espacial.

**Capacidad de procesamiento de imgenes de alta resolucin**: ViT puede tener dificultades para procesar imgenes de alta resolucin debido a la limitacin de memoria y la complejidad computacional. En cambio, Swin Transformer se ha diseado especficamente para abordar esta limitacin mediante el uso de ventanas y reduccin de escala, lo que permite un procesamiento eficiente de imgenes de alta resolucin.

# Datos

```python
data = pd.read_feather("NN_INPUT.feather"); #Dataset

drop=[]
for i,r in data.iterrows():
    if not Path(r["Path"]).exists():
        drop.append(i)

data.drop(drop,inplace=True)


data
```

# Dataset

```python
def GEN():
    for i,r in data.iterrows():
        I,F,E,N=r
        L="UNK"
        if F>0:
            if E>0:
                L="EXP+FUM"
            else:
                L="FUM"
        elif E>0:
            L="EXP"
        elif N>0:
            L="INA"
        yield {
            "image":I,
            "label":L
        }
Data=Dataset.from_generator(GEN).cast_column("image",dsImage()).cast_column("label",ClassLabel(names=["UNK","EXP","FUM","EXP+FUM","INA"]))
```

```python
Data=Data.train_test_split(.05)
```

```python
labels = Data["train"].features["label"].names
label2id, id2label = dict(), dict()
for i, label in enumerate(labels):
    label2id[label] = i
    id2label[i] = label

id2label[2]
```

# Modelo

```python
model_name="microsoft/swinv2-small-patch4-window16-256"
```

```python
image_processor = AutoFeatureExtractor.from_pretrained(model_name)
```

```python
from torchvision.transforms import (
    CenterCrop,
    Compose,
    Normalize,
    RandomHorizontalFlip,
    RandomResizedCrop,
    Resize,
    ToTensor,
)

normalize = Normalize(mean=image_processor.image_mean, std=image_processor.image_std)
if "height" in image_processor.size:
    size = (image_processor.size["height"], image_processor.size["width"])
    crop_size = size
    max_size = None
elif "shortest_edge" in image_processor.size:
    size = image_processor.size["shortest_edge"]
    crop_size = (size, size)
    max_size = image_processor.size.get("longest_edge")

train_transforms = Compose(
        [
            RandomResizedCrop(crop_size),
            RandomHorizontalFlip(),
            ToTensor(),
            normalize,
        ]
    )

val_transforms = Compose(
        [
            Resize(size),
            CenterCrop(crop_size),
            ToTensor(),
            normalize,
        ]
    )

def preprocess_train(example_batch):
    """Apply train_transforms across a batch."""
    example_batch["pixel_values"] = [
        train_transforms(image.convert("RGB")) for image in example_batch["image"]
    ]
    return example_batch

def preprocess_val(example_batch):
    """Apply val_transforms across a batch."""
    example_batch["pixel_values"] = [val_transforms(image.convert("RGB")) for image in example_batch["image"]]
    return example_batch
```

```python
train_ds = Data['train']
val_ds = Data['test']
```

```python
train_ds.set_transform(preprocess_train)
val_ds.set_transform(preprocess_val)
```

```python
val_ds
```

```python
model = AutoModelForImageClassification.from_pretrained(
    model_name, 
    label2id=label2id,
    id2label=id2label,
    ignore_mismatched_sizes = True, # provide this in case you're planning to fine-tune an already fine-tuned checkpoint
)
```

```python
metric = load_metric("accuracy")
```

```python
model_name_b = model_name.split("/")[-1]

args = TrainingArguments(
    f"{model_name_b}-finetuned-popocatepetl",
    remove_unused_columns=False,
    evaluation_strategy = "epoch",
    save_strategy = "epoch",
    learning_rate=5e-5,
    fp16=True,
    per_device_train_batch_size=32,
    #gradient_accumulation_steps=4,
    per_device_eval_batch_size=32,
    num_train_epochs=10,
    warmup_ratio=0.1,
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    push_to_hub=False,
    report_to=[],
)
```

```python
def compute_metrics(eval_pred):
    """Computes accuracy on a batch of predictions"""
    predictions = np.argmax(eval_pred.predictions, axis=1)
    return metric.compute(predictions=predictions, references=eval_pred.label_ids)
```

```python
def collate_fn(examples):
    pixel_values = torch.stack([example["pixel_values"] for example in examples])
    labels = torch.tensor([example["label"] for example in examples])
    return {"pixel_values": pixel_values, "labels": labels}
```

```python
train_ds[0]
```

```python
trainer = Trainer(
    model,
    args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    tokenizer=image_processor,
    compute_metrics=compute_metrics,
    data_collator=collate_fn,
)
```

```python
train_results = trainer.train()
```

```python
# rest is optional but nice to have
trainer.save_model()
trainer.log_metrics("train", train_results.metrics)
trainer.save_metrics("train", train_results.metrics)
trainer.save_state()
```
