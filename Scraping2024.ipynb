```python
import requests
from bs4 import BeautifulSoup

def clean_image_url(url):
    url = url.replace('\\"', '').strip('"')
    if url.endswith('\\'):
        url = url.rstrip('\\')
    return url

image_urls = []

for page in range(1, 31):
    url = f"https://www.gob.mx/cenapred/es/archivo/articulos?filter_origin=archive&idiom=es&order=DESC&page={page}"
    response = requests.get(url)
    page_content = response.content
    
    soup = BeautifulSoup(page_content, 'html5lib')
    image_tags = soup.find_all('img')
    
    for image_tag in image_tags:
        raw_url = image_tag.get('src')
        if raw_url:
            cleaned_url = clean_image_url(raw_url)
            image_urls.append(cleaned_url)
image_urls
```

```python
import requests
import os

# Lista de URLs de imágenes limpias (asegúrate de que esta lista ya esté definida)
# image_urls = [ ... ]

# Crear carpeta para guardar las imágenes si no existe
output_folder = 'images_2024'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Descargar las imágenes
for i, img_url in enumerate(image_urls):
    try:
        # Descargar la imagen
        img_response = requests.get(img_url)
        img_response.raise_for_status()  # Verificar que la solicitud fue exitosa
        
        # Definir el nombre del archivo
        img_name = os.path.join(output_folder, f'image_{i+1}.jpg')
        
        # Guardar la imagen en el archivo
        with open(img_name, 'wb') as f:
            f.write(img_response.content)
        
        print(f'Descargada imagen {i+1}: {img_url}')
    except Exception as e:
        print(f'Error al descargar la imagen {i+1}: {e}')

print(f'Total de imágenes descargadas: {len(image_urls)}')
```
